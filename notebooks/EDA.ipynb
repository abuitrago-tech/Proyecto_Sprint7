## Entendimiento de los datos

```python
# Cargar librerias
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px

import plotly.io as pio
pio.templates.default = "plotly"
```

```python
# Cargar dataset
car_data = pd.read_csv(r'C:\Users\user\Analisis_de_datos\Sprint_7\Directorio_Principal\Proyecto_Sprint7\vehicles_us.csv')
car_data["type"] = car_data["type"].astype(str)
```

```python
car_data.info()
```

```python
car_data['Fabricante'] = car_data['model'].str.split(' ').str[0].str.split('-').str[0]
car_data.head()
```

```python
car_data.sample(10)
```

```python
car_data.describe().round(1)
```

## Visualizaciones

```python
# Visualizar histograma 1
fig_fabricantes = px.histogram(
    car_data, 
    x= "Fabricante", 
    color="type",  
    width = 700, 
    height = 600, 
    title= f"Tipo de veh√≠culos por fabricante"
) 
fig_fabricantes.show()
```

```python
# Visualizar histograma 2
fig_model_year = px.histogram(
    car_data, 
    x= "model_year", 
    color="condition",  
    width = 700, 
    height = 600, 
    title= "Condici√≥n de Veh√≠culos"
) 
fig_model_year.show()
```

```python
# Compraci√≥n de dos variables
opciones = list(car_data.columns)[0:14]
v = (opciones[0],opciones[1])
```

```python
# Calcular estadisticos de price
prom = np.mean(car_data[v[0]])
med = np.median(car_data[v[0]])
desv = np.std(car_data[v[0]])
print("Media: {:.1f}".format(prom))
print("Mediana: {:.1f}".format(med))
print("Desviaci√≥n: {:.1f}".format(desv))
```

```python
# Visualizar dispersion entre variables
fig_price = px.scatter(
    car_data, 
    x = v[1], 
    y = v[0], 
    color = "type", 
    title = f"Dispersi√≥n {v[0]} vs. {v[1]}",
    width = 700, 
    height = 400
) 
fig_price.show()
```
